```python
import pandas as pd
import sqlite3

products = pd.read_csv('C:/Users/Mayara Lopes/Desktop/crud_python/static/data_woo.csv')
conn = sqlite3.connect('data.db')
c = conn.cursor()
```

```python
products = products[[
    'Name', 'ID', 'Type', 'Description',
    'Stock', 'Sale price', 'Tax status',
]]
```

```python
products = products.rename(columns={
    'Name': 'name', 'ID': 'id', 'Type': 'type', 'Description': 'description', 
    'Stock': 'stock', 'Sale price': 'sale_price', 'Tax status': 'tax_status'
    })
```

```python
c.execute('''CREATE TABLE products (id int, name text, type text, description text, stock int, sale_price real, tax_status text)''')
```

```python
products.to_sql('products', conn, if_exists='append', index = False)
```

```python
c.execute('''SELECT * FROM products''').fetchall()
```
